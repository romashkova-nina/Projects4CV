## Задание 5 (30 баллов)

По БД sample_analytics выведите по каждому customer_id (выведите также username аккаунта):

* Сумма лимитов у пользователя (сумма по всем его аккаунтам)

* Общее число транзакций пользователя (сумма по всем аккаунтам)

* (на еще +20 баллов) добавить сумму всех пополнений и трат по транзакциям

```python
db5 = cluster["sample_analytics"]
collection51=db5["customers"]
collection52=db5["accounts"]
collection53=db5["transactions"]
```

```python
#Найдем сумму лимитов и число транзакций у каждого пользователя
res51 = collection51.aggregate([ 
        {"$unwind": "$accounts"},
        {"$lookup": {"from": "accounts", "localField": "accounts", "foreignField": "account_id", "as": "acc_info"}}, 
        {"$lookup":{"from": "transactions", "localField": "accounts", "foreignField": "account_id", "as": "transac_info"}},
        {"$unwind": "$acc_info"},
        {"$unwind": "$transac_info"},
        {"$group": {"_id":{"_id":"$_id", "username":"$username"}, "limits_total":{"$sum":"$acc_info.limit"},
                   "transactions_total":{"$sum":"$transac_info.transaction_count"}}},
        {"$project":{"_id":0, "customer_id":"$_id._id", "username":"$_id.username", "limits_total":"$limits_total", "transactions_total":"$transactions_total"}}
        
])

for i in res51:
    print(i)
```

```python
#Для каждого customer_id создадим 3 поля: сумму трат (total_spent), сумму пополнений (total_earned)
#и баланс (total_earned-total_spent) по всем транзакциям
res52 = collection51.aggregate([ 
        {"$unwind": "$accounts"},
        {"$lookup": {"from": "accounts", "localField": "accounts", "foreignField": "account_id", "as": "acc_info"}}, 
        {"$lookup":{"from": "transactions", "localField": "accounts", "foreignField": "account_id", "as": "transac_info"}},
        {"$unwind": "$acc_info"},
        {"$unwind": "$transac_info"},
        {"$unwind": "$transac_info.transactions"},
        {"$addFields": {"total_transac": {"$toDouble": "$transac_info.transactions.total"}}},
        {"$addFields": {"spent": {"$cond":[{"$eq": ["$transac_info.transactions.transaction_code", "buy"]}, "$total_transac", 0]},
                       "earned": {"$cond":[{"$eq": ["$transac_info.transactions.transaction_code", "sell"]}, "$total_transac", 0]}}},
        {"$group":{"_id":{"customer_id":"$_id", "username":"$username"}, "total_spent":{"$sum":"$spent"}, "total_earned":{"$sum":"$earned"}}},      
        {"$addFields": {"balance": {"$subtract": ["$total_earned", "$total_spent"]}}},
        {"$project": {"customer_id":"$_id.customer_id", "username":"$_id.username", "total_spent":"$total_spent",
                     "total_earned":"$total_earned", "balance":"$balance", "_id":0}}
])

for i in res52:
    print(i)   
```
